```julia
using Statistics, Plots
```

```julia
function pr(alpha, beta)
    0.25 * binomial(2, alpha)*(0.25 + 0.9^alpha*0.1^(2-alpha)) *
        binomial(2, beta)*(0.25 + 0.9^beta*0.1^(2-beta))
end
```

```julia
pr(0,0)
```

```julia
sum = 0
for i in 0:2
    for j in 0:2
        sum += pr(i,j)
    end
end
sum
```

When we have HH, HT then the likelihood is:

```julia
function likelihood(theta1, theta2)
    theta1^2*theta2*(1-theta2)
end
```

```julia
theta1 = [0.5, 0.9]
theta2 = [0.5, 0.9]

surface(theta1, theta2, likelihood, xlabel="Theta1", ylabel="Theta2", zlabel="Likelihood")
```

```julia
marginal_likelihood = 0.25*(0.5^2+0.9^2)*(0.5^2+0.9*0.1)
```

```julia
function posterior(theta1, theta2)
    0.25/0.0901 * theta1^2*theta2*(1-theta2)
end
```

```julia
surface(theta1, theta2, posterior, xlabel="Theta1", ylabel="Theta2", zlabel="Posterior")
```

```julia
expectation_theta_1 = 0
for theta_1 in [0.5, 0.9]
    for theta_2 in [0.5, 0.9]
        expectation_theta_1 += theta_1*posterior(theta_1, theta_2)
    end
end
```
